```python
df=pd.read_csv('stock.csv')
df
```

1. a) histogram

```python
plt.figure(figsize=(10,27))


plt.subplot(3,1,1)
plt.title('histogram Purchase Price')
plt.hist(df['Purchase Price'], color='r')
plt.xlabel('Purchase Price', fontsize=8)
plt.ylabel('Frekuensi', fontsize=8)
plt.xticks(rotation=45)

plt.subplot(3,1,2)
plt.title('histogram Current Price')
plt.hist(df['Current Price'], color='g')
plt.xlabel('Current Price', fontsize=8)
plt.ylabel('Frekuensi', fontsize=8)
plt.xticks(rotation=90)

plt.subplot(3,1,3)
plt.title('histogram Gain/Loss')
plt.hist(df['Gain/Loss'], color='y')
plt.xlabel('Gain/Loss', fontsize=8)
plt.ylabel('Frekuensi', fontsize=8)
plt.xticks(rotation=90)

plt.show()
```
